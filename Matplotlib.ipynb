```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [1,4,9,16]

plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Graph")
plt.show()
```

```python
category = ["A","B","C"]
values = [10,20,15]

plt.plot(category, values)
plt.xlabel(category)
plt.ylabel(values)
plt.title("Bar Chart")
plt.show()
```
